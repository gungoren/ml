```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = ["1. USD", "2. CHINA", "3. JAPANESE", "4. GERMANY", "5. ENGLAND", "6. FRANCE", "7. TURKEY"]
y = [20410, 14090, 5170, 4210, 2940, 2930, 3231]
z = [63039, 9957, 40649, 51158, 44162, 44915, 32332]
```

```python
plt.plot(x,y, "r")
plt.xlabel("Countries")
plt.ylabel("GPD")
plt.title("GPD by Countries")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, "r")
plt.subplot(1,2,2)
plt.plot(x,z, "y")
plt.tight_layout()
```

```python
f = plt.figure()
axes = f.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x,y)
axes.set_xlabel("Countries")
axes.set_ylabel("Gpd")
axes.set_title("Title")
```

```python
f = plt.figure()
axes1 = f.add_axes([0.1, 0.1, 0.9,0.9])
axes2 = f.add_axes([0.5, 0.5, 0.4,0.4])
axes1.plot(x, y, "y")
axes2.plot(x, z, "r")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,"b")
axes[0].set_title("Gpd")

axes[1].plot(x,z,"y")
axes[1].set_title("Gpd by Person")

plt.tight_layout()
```
