```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = ["1. USD", "2. CHINA", "3. JAPANESE", "4. GERMANY", "5. ENGLAND", "6. FRANCE", "7. TURKEY"]
y = [20410, 14090, 5170, 4210, 2940, 2930, 3231]
z = [63039, 9957, 40649, 51158, 44162, 44915, 32332]

f = plt.figure(figsize=[10,6])
axes = f.add_axes([0.0, 0.0, 1.0, 1.0])
axes.plot(x,y)
axes.set_xlabel("Countries")
axes.set_ylabel("Gpd")
axes.set_title("Title")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[13,6])
axes[0].plot(x,y,"b")
axes[0].set_title("Gpd")

axes[1].plot(x,z,"y")
axes[1].set_title("Gpd by Person")
```

```python
fig.savefig("../../data/fig.png")
```

```python
fig = plt.figure(figsize=(10,5))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, "r", label="GPD")
axes.plot(x, z, "y", label="shared GPD")
axes.legend(loc=5)
```
