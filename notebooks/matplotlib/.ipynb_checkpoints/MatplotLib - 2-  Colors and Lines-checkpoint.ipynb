```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = ["1. USD", "2. CHINA", "3. JAPANESE", "4. GERMANY", "5. ENGLAND", "6. FRANCE", "7. TURKEY"]
y = [20410, 14090, 5170, 4210, 2940, 2930, 3231]
z = [63039, 9957, 40649, 51158, 44162, 44915, 32332]
```

```python
f = plt.figure()
axes = f.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x,y, color="r", linewidth=1
          , linestyle="-.", marker="o", markersize=20,
         markerfacecolor="y", markeredgewidth=4, 
          markeredgecolor="b",
         alpha=0.5)
axes.set_xlabel("Countries")
axes.set_ylabel("Gpd")
axes.set_title("Title")
```

```python
f = plt.figure()
axes = f.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x,y, color="r", linewidth=1
          , linestyle="-.")
axes.set_xlabel("Countries")
axes.set_ylabel("Gpd")
axes.set_title("Title")

axes.set_xlim(2,6)
axes.set_ylim(2000,7500)
```
