```python
df = pd.read_csv("../../data/seaborn/dünyam.csv")
df.dropna(how="any", inplace=True)
df
```

```python
df.dropna(how="any", inplace=True)
```

```python
sns.set(style="dark", palette="muted", font_scale=1.5)
sns.distplot(df["Doğum oranı"], bins=20, kde=False, color="y")
```

```python
sns.set(style="darkgrid", palette="muted", font_scale=1.2)
sns.distplot(df["Yıllık değişim oranı"], color="r", hist=False, rug=True)
```

```python
sns.set(style="white", palette="Blues", font_scale=1.2)
sns.distplot(df["Ortalama yaş"], color="g", hist=False, kde_kws={"shade":True})
```

```python
sns.set(style="whitegrid", palette="Blues", font_scale=1.2)
sns.distplot(df["Nüfüs"], color="m")
```

```python
sns.set(style="darkgrid", font_scale=1.5)
f, axes = plt.subplots(2, 2, figsize=[15,10])
sns.distplot(df["Doğum oranı"], bins=20, kde=False, color="y", ax = axes[0,0])
sns.distplot(df["Yıllık değişim oranı"], color="r", hist=False, rug=True, ax = axes[0,1])
sns.distplot(df["Ortalama yaş"], color="g", hist=False, kde_kws={"shade":True}, ax=axes[1,0])
sns.distplot(df["Nüfüs"], color="m", ax=axes[1,1])
plt.tight_layout()
```

### pairgrid()
